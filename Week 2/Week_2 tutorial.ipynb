```python
def compare(x, y):
    """
    This is a function to compare two numbers, x and y
    """
    if x > y:
        print(f"{x} is greater than {y}")
    elif x < y:
        print(f"{x} is smaller than {y}")
    else:
        print(f"{x} is equal to {y}")
```

```python
import matplotlib.pyplot as plt
```

```python
black_img = np.zeros((200,200), dtype=np.uint8)

plt.imshow(black_img, cmap=plt.cm.gray)
plt.title("black grayscale")
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
white_img = 255 * np.ones((200, 200), dtype=np.uint8)

plt.imshow(white_img, cmap=plt.cm.gray, vmin=0, vmax=256)
plt.title("white grayscale")
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# create a random color image
random_color_img = np.random.randint(0, high=256, size=(200,200, 3), dtype=np.uint8)

plt.imshow(random_color_img)
plt.title("Random color image")
plt.xticks([]), plt.yticks([])
plt.show()
```

## Exercise 1

```python
def integerOddEven(x):
    """
    This function is to find out if the integer is odd or even
    """
    
    if isinstance(x , int):
        if x%2 == 0:
            if x == 0:
                statement = " is neither odd or even number."
            else:
                statement = " is even number."
        else:
            statement = " is odd number."
    else:
        statement = " is not an integer!"
        
    return print(f"{x}" + statement)
```

```python
integerOddEven(4)
integerOddEven(9)
integerOddEven(0)
integerOddEven(1.1)
```

## Exercise 4

```python
# Create black image w/ dimension of 200x200
black_img = np.zeros((200, 200), dtype=np.uint8)

# Find out the center coordinate
img_center_x = black_img.shape[1] // 2
img_center_y = black_img.shape[0] // 2


# Define the radius of the circular spot
radius = 35

# Create a meshgrid of x and y values
x, y = np.meshgrid(np.arange(black_img.shape[1]), np.arange(black_img.shape[0]))

# Calculate the distance from each pixel to the center
distances = np.sqrt((x - img_center_x)**2 + (y - img_center_y)**2)

# Change the circle to white color
black_img[distances <= radius] = 255


plt.imshow(black_img, cmap='gray')
plt.title("Black image with white circular spot")
plt.xticks([]), plt.yticks([])
plt.show()
```
